# Use different visualization like countplot & scatterplots.

**1. Scatter Plot to Correlate Parental level of Education and Reading Score.**

```
studentData.plot.scatter("parental level of education","reading score",figsize=(10,3))
```

* From the above scatter plot we could see that the students whose parents have got bachelor's or master's degree have a reading score above 30.
* Whereas the students having reading score below 30 are almost the children of parents who has got the educational level of some high school,high school and some college.






**2. Count Plot**

* Now by using the help of a countplot we can easily get the count of students with reading score below 30 and whose parents are having a education level of High school,some high school and some college.

```
std= studentData[studentData["reading score"]<30]
```

```
import seaborn as sns
sns.countplot(std["parental level of education"],orient="h")
```

* Now, from the above countplot we can clearly get the exact number of students with reading score below 30 and whose parents are having education level of some high school, high school and some college.
   * Number of students with reading score below 30(parental level of education- some high school) = 2
   * Number of students with reading score below 30(parental level of education- some colllege) = 2
   * Number of students with reading score below 30(parental level of education- High school) = 4
